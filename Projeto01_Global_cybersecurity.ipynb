# Este dataset é acerca ameaças da cybersegurança globais, suas respectivas causas com o passar do tempo, mais especificamente no período de tempo de 2014 a 2024, ou seja, 10 anos

# Com base na visualização de dados acima, temos suas respectivas colunas cujo significado são:

Country -> País que aconteceu o ataque;

Year -> O Ano que aconteceu o ataque de falha de segurança;

Attack Type -> O tipo de ataque/violação de cybersecurity

Target Industry -> Os Alvos Empresariais, isto é, em que local o ataque foi ocasionado;

Financial Loss (in Million $) -> Perda financeira em Milhões de Dólares;

Number of Affected Users -> Número de usuários afetados com os respectivos ataques;

Attack Source -> Fonte de ataques

Security Vulnerability Type -> Tipo de Vulnerabilidade de segurança;

Defense Mechanism Used -> Mecanismo de defesa utilizado;

Incident Resolution Time (in Hours) -> Tempo de resolução do incidente (em horas)

```python
media = df.loc[:, 'Number of Affected Users'].mean()
mean = df['Incident Resolution Time (in Hours)'].mean()
print(media)
print(mean)

# calculando e mostrando as respectivas médias aritiméticas de todas as colunas (somente colunas que contém número)
```

```python
maximo = df.loc[:, 'Number of Affected Users'].max()
maximo_tempo_resolucao = df['Incident Resolution Time (in Hours)'].max()
print(maximo)
print(maximo_tempo_resolucao)

# calculando e mostrando o máximo de todas as colunas (somente colunas que contém número)
minimo = df.loc[:, 'Number of Affected Users'].min()
minimo_tempo_resolucao = df['Incident Resolution Time (in Hours)'].min()
print(minimo_tempo_resolucao)
print(minimo)

# calculando e mostrando o máximo e o mínimo de todas as colunas (somente colunas que contém número)
```

```python
desvio = df.loc[:, 'Number of Affected Users'].std()
print(desvio)

# calculando e mostrando o desvio padrão de todas as colunas (somente colunas que contém número)
```

```python
import matplotlib.pyplot as plt

plt.barh(df['Year'], df['Number of Affected Users'], )
plt.xlabel('Ano', fontsize = 12)
plt.ylabel('Número de Usuários Afetados',  fontsize = 12)
# plt.yscale('log')
# plt.ylim((df['Number of Affected Users'].min(), df['Number of Affected Users'].max()))
plt.xticks(rotation=90, ha='right')
plt.title('Número de Usuários Afetados por Ano')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()

# analisando os dados de quaantidade de usuários afetados por ano
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Country'], df['Number of Affected Users'], color='orange')
plt.xlabel('País', fontsize = 12)
plt.ylabel('Número de Usuários Afetados',  fontsize = 12)
plt.xticks(rotation=90, ha='right')
plt.title('Número de Usuários Afetados por País')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()

# analisando os dados de quaantidade de usuários afetados por ano
```

```python
import matplotlib.pyplot as plt

plt.barh(df['Attack Type'], df['Year'], color='red')
plt.xlabel('Número de Usuários Afetados', fontsize = 12)
plt.ylabel('Tipo de ataque',  fontsize = 12)
plt.xticks(rotation=90, ha='right')
plt.title('Tipos de ataques cibernéticos por números de usuários')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```
